```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
df=pd.read_csv("adult.csv", sep=',', names=['age','workclass','fnlwgt','education','educationNum','maritalStatus','occupation','relationship','race','sex','capitalGain','capitalLoss','hoursPweek','nativeCountry','income'])
df.head(15)
```

```python
df.income = df.income.replace('<=50K',0)
df.income = df.income.replace('>50K',1)
df.describe()
```

```python
df.info()
```

```python
df.describe(include='object')
```

```python
#Проверим есть ли в датасете строки и столбцы с нулевыми значениями 
df[df.isnull().values==True]
```

```python
#Взглнем на уникальные значения категориальных характеристик 
print('workclass',df.workclass.unique())
print('education',df.education.unique())
print('maritalStatus',df['maritalStatus'].unique())
print('occupation',df.occupation.unique())
print('relationship',df.relationship.unique())
print('race',df.race.unique())
print('sex',df.sex.unique())
print('nativeCountry',df['nativeCountry'].unique())

```

```python
#Заменим вопр знаки у категориальных характеристик, в которых он встречается
df['workclass']=df['workclass'].str.replace('?','Private',regex=True)
df['occupation']=df['occupation'].str.replace('?','Prof-specialty',regex=True)
df['nativeCountry']=df['nativeCountry'].str.replace('?','United-States',regex=True)
```

```python
#Осуществим группировку некоторых значений категориальных характеристик
df['maritalStatus']= df['maritalStatus'].replace(tuple(['Married-civ-spouse', 'Married-AF-spouse']), 'married',regex=True)
df['maritalStatus']= df['maritalStatus'].replace(tuple(['Never-married']), 'not-married',regex=True)
df['maritalStatus']= df['maritalStatus'].replace(tuple(['Divorced', 'Separated','Widowed','Married-spouse-absent']), 'Other',regex=True)
```

```python
df.education= df.education.replace(['Preschool', '1st-4th', '5th-6th', '7th-8th', '9th','10th', '11th', '12th'], 'Basic edu',regex=True)
df.education = df.education.replace(['Assoc-voc', 'Assoc-acdm', 'Prof-school', 'Some-college'], 'Middle edu', regex=True)
df.education.replace('HS-grad', 'HighGrad',inplace=True)
df.education.replace('Prof-school', 'Masters',regex=True)

df.education.unique()
```

```python
# Также преобразуем характеристику income в числовую
df.income = df.income.factorize()[0]
```

Label кодирование меток перед обучением


```python
df1= df.copy()
df1= df1.apply(LabelEncoder().fit_transform)
df1.head()
```

```python
X= df.drop(['income'], axis=1)
y = df['income']
```

Масштабируем данные

```python
ss= StandardScaler().fit(df1.drop('income', axis=1))
X= ss.transform(df1.drop('income', axis=1))

y= df['income']
# разделим наши данные на обучающий и тестовый наборы.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=30)
```

КЛАССИФИКАЦИОННЫЕ МОДЕЛИ


Logical Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr = LogisticRegression()

lr.fit(X_train, y_train)
score_lr=lr.score(X_test, y_test)

print("Accuracy on test data: {:,.3f}".format(score_lr))
print("Cross val: {:,.3f}".format(np.mean(cross_val_score(lr,X,y,cv=4))))
```

Linear Support Vector Machine

```python
from sklearn.svm import SVC
svc = SVC(gamma=0.2)
svc.fit(X_train, y_train)
score_svc = svc.score(X_test,y_test)
print("Accuracy on test data: {:,.3f}".format(score_svc))
np.mean(cross_val_score(svc,X,y,cv=4))
```

Random Forest


```python
from sklearn.ensemble import RandomForestClassifier
#задаем гиперпараметры
rfc = RandomForestClassifier(n_estimators = 200, max_depth=20, random_state=30)
rfc.fit(X_train,y_train)
score_rfc=rfc.score(X_test, y_test)
print("Accuracy on test data: {:,.3f}".format(score_rfc))
print("Cross val: {:,.3f}".format(np.mean(cross_val_score(rfc,X,y,cv=4))))
```

Оценка метрики accurancy

```python
#Лучшее значение точности достиглось в случае использ-я классификатора RFC. Получим оценку метрики
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 30)
rfc.fit(X_train,y_train)
y_pred = rfc.predict(X_test)
print("Оценка метрики: {:,.3f}".format(accuracy_score(y_test, y_pred)))
```
